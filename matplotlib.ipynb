# matplotlib

## Bar grapgh Vertical

```python
from matplotlib import pyplot as plt

x=[1,2,3,4,5]
y=[10,20,10,30,20]

plt.xlabel("students")
plt.ylabel("marks")
 
c=["green","blue","red","orange","pink"]
plt.bar(x,y,color=c,width=0.5)
plt.title("BAR CHART")
 

plt.show()
```

# combine 2 - bar graph

```python
from matplotlib import pyplot as plt
x = [5,8,10]
y = [12,16,6]
x2 = [6,9,11]
y2 = [6,15,7]
plt.bar(x, y, align = 'center',label='x,y')
plt.bar(x2, y2, color = 'g', align = 'center',label='x2,y2')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.legend(title="data")
plt.show()
```

# Histograph

```python
from matplotlib import pyplot as plt
students=[1,10,20,35,50,90,83,85,13,48,55]
age_intervals=[10,35,70,100]
plt.xlabel("age _intervals")
plt.ylabel(" no of students")
plt.hist(students,age_intervals)
plt.show
```

```python
from matplotlib import pyplot as plt
import numpy as np
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a, bins = [0,20,40,60,80,100])
plt.title("histogram")
plt.show()
```

```python
from matplotlib import pyplot as plt

movie=[1,2,3,4]
rate=[500,300,100,400]

 
c=["red","green","yellow","orange"]

plt.pie(rate,colors=c,labels=movie,shadow=True,explode=[0.2,0,0,0])
plt.legend(title="rate movie")
plt.show()
```

## BAR Graph Horizontal

```python
from matplotlib import pyplot as plt

x=[1,2,3,4,5]
y=[10,20,10,30,20]

plt.xlabel("students")
plt.ylabel("marks")

c=["green","blue","red","orange","pink"]
plt.barh(x,y,color=c,height=0.5)
plt.title("BAR CHART")

plt.show()
```

# Pie Chart

```python
from matplotlib import pyplot as plt

branch=["cse","ece","eee","mech","civil"]
strength=[300,170,250,100,200]

c=["green","red","orange","pink","blue"]

plt.pie(strength,colors=c,explode=[0.2,0,0,0,0],shadow=True,labels=branch) #shadow T/F Explode - 0 default incre. to seperate
plt.legend(title="Branch") # legend -  key point display 
plt.title("OverView")
plt.show()
```

## subplot

```python
from matplotlib import pyplot as plt

x=[0,1,2,3]
y=[3,6,2,9]

plt.subplot(1,4,1)
plt.plot(x,y,linestyle="dashed",color="red")
plt.title("subplot-1")

y1=[0,1,2,3]
x1=[3,6,2,9]

plt.subplot(1,4,2)
plt.plot(x1,y1,linestyle="dotted",color="green",linewidth=1.5)

plt.title("subplot-2")

y2=[0,1,2,3]
x2=[3,6,2,9]

plt.subplot(1,4,3)
plt.plot(x2,y2,linestyle="dotted",color="green",linewidth=1.5)

plt.title("subplot-3")

y3=[0,1,2,3]
x3=[3,6,2,9]

plt.subplot(1,4,4)
plt.plot(x3,y3,linestyle="dotted",color="green",linewidth=1.5)

plt.title("subplot-4")

plt.show()
```
