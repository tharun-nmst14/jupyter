# matplotlib

## Bar grapgh Vertical

```python
from matplotlib import pyplot as plt

x=[1,2,3,4,5]
y=[10,20,10,30,20]

plt.xlabel("students")
plt.ylabel("marks")
 
c=["green","blue","red","orange","pink"]
plt.bar(x,y,color=c,width=0.5)
plt.title("BAR CHART")
 

plt.show()
```

```python
from matplotlib import pyplot as plt

movie=[1,2,3,4]
rate=[500,300,100,400]

 
c=["red","green","yellow","orange"]

plt.pie(rate,colors=c,labels=movie,shadow=True,explode=[0.2,0,0,0])
plt.legend(title="rate movie")
plt.show()
```

## BAR Graph Horizontal

```python
from matplotlib import pyplot as plt

x=[1,2,3,4,5]
y=[10,20,10,30,20]

plt.xlabel("students")
plt.ylabel("marks")

c=["green","blue","red","orange","pink"]
plt.barh(x,y,color=c,height=0.5)
plt.title("BAR CHART")

plt.show()
```

# Pie Chart

```python
from matplotlib import pyplot as plt

branch=["cse","ece","eee","mech","civil"]
strength=[300,170,250,100,200]

c=["green","red","orange","pink","blue"]

plt.pie(strength,colors=c,explode=[0.2,0,0,0,0],shadow=True,labels=branch) #shadow T/F Explode - 0 default incre. to seperate
plt.legend(title="Branch") # legend -  key point display 
plt.title("OverView")
plt.show()
```

# Line Style

## plot - line grapgh / dashed /dotted /

```python
from matplotlib import pyplot as plt
x=[1,4,1,2,3]
y=[2,4,6,1,5]
z=[5,3,1,6,2]
plt.xlabel("x")
plt.ylabel("y")

plt.plot(x,linewidth=0.5,linestyle="dotted",color="red")
plt.plot(y,linewidth=1.5,linestyle="dashed",color="green")
plt.plot(z,linewidth=1.5,color="yellow")
    # plt.plot(x,z,color="green",linewidth=1) # combine x,y with same size
    
    #plt.legend(title=["x","y"])
#k=input("enter var to legend")
#plt.legend(title=k)

plt.title("line Style")
plt.show()
```

# Dot graph
## dotted  - scatter

```python
from matplotlib import pyplot as plt

x=[3,2,6,5,1]
y=[6,3,2,4,5]
s=[100,300,200,150,100]
plt.scatter(x,y,linewidth=0,sizes=s)
plt.show()
```

## subplot

```python
from matplotlib import pyplot as plt

x=[0,1,2,3]
y=[3,6,2,9]

plt.subplot(1,4,1)
plt.plot(x,y,linestyle="dashed",color="red")
plt.title("subplot-1")

y1=[0,1,2,3]
x1=[3,6,2,9]

plt.subplot(1,4,2)
plt.plot(x1,y1,linestyle="dotted",color="green",linewidth=1.5)

plt.title("subplot-2")

y2=[0,1,2,3]
x2=[3,6,2,9]

plt.subplot(1,4,3)
plt.plot(x2,y2,linestyle="dotted",color="green",linewidth=1.5)

plt.title("subplot-3")

y3=[0,1,2,3]
x3=[3,6,2,9]

plt.subplot(1,4,4)
plt.plot(x3,y3,linestyle="dotted",color="green",linewidth=1.5)

plt.title("subplot-4")

plt.show()
```
