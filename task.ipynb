## Sem-Cgpa Bargrph

```python
from matplotlib import pyplot as plt
sem=[1,2,3]
cgpa=[8.7,8.6,8.4]

plt.xlabel("sem")
plt.ylabel("cgpa")
c=["pink","yellow","green"]
plt.bar(sem,cgpa,width=0.3,color=c)
 
plt.title("GPA")
plt.show()
```

## line - Maths Vs science

```python
from matplotlib import pyplot as plt
maths=[50,62,43]
science=[60,57,35]

 
plt.plot(maths,color="green")
plt.plot(science,color="red")
plt.title("maths vs science")
plt.show()
 
```

# pie  - time spent

```python
from matplotlib import pyplot as plt
activity=["sleep","work","study","roam","social"]
time=[80,24,80,40,30]

c=["green","red","orange","pink","black","yellow"]

plt.pie(time,colors=c,shadow=True,labels=activity,explode=[0.1,0,0,0,0])
plt.legend(title="spent")
plt
plt.show()
```

## subplot maths Vs science

```python
maths=[1,2,3,4]
mmarks=[50,70,90,100]
science=[1,2,3,4]
smarks=[20,40,80,100]


plt.subplot(1,2,1)
plt.plot(maths,mmarks,linestyle="dotted",color="green",linewidth=1.5)

plt.title("Maths")

plt.subplot(1,2,2)
plt.plot(science,smarks,linestyle="dotted",color="red",linewidth=1.5)

plt.title("Science")

plt.show()
```
